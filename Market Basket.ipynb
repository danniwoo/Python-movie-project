```python
Revenue=data.copy()
```

```python
Revenue=Revenue.drop(Revenue[Revenue['Price']=='0'].index)
Revenue=Revenue.drop(Revenue[Revenue['Itemname']=='Adjust bad debt'].index)
```

```python
New=Revenue.copy()
New['Total']=New['Quantity']*New['Price']
```

```python
New=New.drop(columns=['Quantity','Price'])
New
```

```python
shopping = New

# Group by 'BillNo' and aggregate values
grouped = New.groupby('BillNo').agg({
    'Itemname': lambda x: ', '.join(x),
    'Date': 'first',
    'CustomerID': 'first',
    'Country': 'first',
    'Total': 'sum'
}).reset_index()

# Assign columns to the 'grouped' DataFrame
grouped.columns = ['BillNo', 'Itemname', 'Date', 'CustomerID', 'Country', 'Total']

# Display the grouped DataFrame
grouped
```
